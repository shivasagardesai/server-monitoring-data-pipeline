```python
df["Log_Timestamp"] = pd.to_datetime(df["Log_Timestamp"], dayfirst=True)
```

```python
df.info()
```

Data Transformation

```python
df["Total_Network_Traffic"] = df["Network_Traffic_In (MB/s)"] + df["Network_Traffic_Out (MB/s)"]
```

```python
df.head(5)
```

```python
df["Availability_Rate"] = (df["Uptime (Hours)"] /(df["Uptime (Hours)"] + df["Downtime (Hours)"])) * 100
```

```python
df.head(2)
```

```python
df["Server_Health_Score"] = ( 100 - (df["CPU_Utilization (%)"] * 0.4) - (df["Memory_Usage (%)"] * 0.3) - (df["Disk_IO (%)"] * 0.3))
```

```python
df.head(5)
```

Export New Dataset

```python
df.to_csv("new_server_logs.csv", index=False)
```
